Diagnostic Functions For GK3D Particle Sim. 

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

im = 8
jm = 8 #temporary
km = 8
i=0
data = np.zeros(im*jm*km)
den = np.zeros((im, jm, km))
den_file = open("den.csv", "r")
for line in den_file:
    data[i] = line
    i = i+1
den = np.array(data).reshape((im,jm,km))
```

```python
#Code to read particle.csv | 
import numpy as np
import matplotlib.pyplot as plt

mm = 1000 #temporary

x = np.zeros(mm)
y = np.zeros(mm)
z = np.zeros(mm)
vpar = np.zeros(mm)
mu = np.zeros(mm)
i=0
check = 0
particle_file = open("particle.dat", "r")
for line in particle_file:
    if check == 1:
        line = line.split()
        x[i] = line[0]
        y[i] = line[1]
        z[i] = line[2]
        vpar[i] = line[3]
        mu[i] = line[4]
        i = i+1
    check = 1
```

```python
den2d=np.zeros((im,jm))
kmhalf=int(km/2)-1
den2d[:,:]=den[:,:,kmhalf]
den1d=np.zeros(im)
jmhalf=int(jm/2)-1
den1d[:]=den2d[:,jmhalf]
```

```python
# Here is an example of how to use contourf(),
# xlim() and ylim().

f=np.zeros((101,101))
x=np.linspace(0,im,im)
y=np.linspace(0,jm,jm)
fTranspose=np.transpose(den2d)
plt.contourf(x,y,fTranspose,20, cmap='GnBu')
plt.colorbar()
plt.title('den(x,y,z=lz/2)')
plt.xlabel('x')
plt.ylabel('y')
#plt.xlim(2.,3.)
#plt.ylim(0.,2.)
plt.show()
```

```python
plt.plot(x,den1d)
plt.xlabel('x')
plt.ylabel('den(x,y=ly/2,z=lz/2)')
plt.show()
```

```python
den2d=np.zeros((im,km))
jmhalf=int(jm/2)-1
den2d[:,:]=den[:,jmhalf,:]
den1d=np.zeros(im)
kmhalf=int(km/2)-1
den1d[:]=den2d[:,kmhalf]
```

```python
# Here is an example of how to use contourf(),
# xlim() and ylim().

f=np.zeros((101,101))
x=np.linspace(0,im,im)
z=np.linspace(0,km,km)
fTranspose=np.transpose(den2d)
plt.contourf(x,z,fTranspose,20, cmap='GnBu')
plt.colorbar()
plt.title('den(x,y=ly/2,z)')
plt.xlabel('x')
plt.ylabel('z')
#plt.xlim(2.,3.)
#plt.ylim(0.,2.)
plt.show()
```

```python
plt.plot(x,den1d)
plt.xlabel('x')
plt.ylabel('den(x,y=ly/2,z=lz/2)')
plt.show()
```
